```python
anime_df = df.pivot(index='anime_id',
                    columns='name',
                    values='rating')
anime_df
```

```python
#Impute those NaNs with 0 values
anime_df.fillna(0, inplace = True)
anime_df
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_similar = 1 - pairwise_distances( anime_df.values,metric='cosine')

# you get cos(theta)
```

```python
user_similar
```

```python
np.fill_diagonal(user_similar , 0)

# fills diagonal as 0 cuz the user with itself will always be similar i.e cos 0 = 1 so we dont want these in our model
# so we replace it with 0 to get better result
```

```python
#Store the results in a dataframe
user_similar_df = pd.DataFrame(user_similar)  
```

```python
user_similar_df
```

```python
#Set the index and column names to user ids 
user_similar_df.index = df['anime_id'].unique()
user_similar_df.columns = df['anime_id'].unique()

user_similar_df
```

```python
#Considering the most similar customer
def recommend (custID):
    simID= user_similar_df.loc[custID].idxmax()  # this will give similar user id AND my given customer ID
    simID_movies =  df[df['anime_id'] == simID].name #this will give list of the movies watched by similar user
    custID_movies = df[df['anime_id'] == custID].name #this will give list of the movies watched by our given user use
    return set (simID_movies) - set (custID_movies) 
```
